```python
import pandas as pd
import mplfinance as mpf
import ta  # Technical Analysis Library


def calculate_atr(df, period=14):
    """Calculate ATR using the True Range formula."""
    df['ATR'] = ta.volatility.average_true_range(df['high'], df['low'], df['close'], window=period)
    return df


if __name__ == '__main__':
    # âœ… Load CSV Data
    df = pd.read_csv('data.csv')

    # âœ… Convert Timestamp to Datetime Format
    df['time'] = pd.to_datetime(df['Timestamp'])
    df.set_index('time', inplace=True)  # Required for mplfinance

    # âœ… Rename columns for mplfinance
    df.rename(columns={'Open': 'open', 'High': 'high', 'Low': 'low', 'Close': 'close', 'Volume': 'volume'}, inplace=True)

    # âœ… Calculate ATR
    df = calculate_atr(df, period=14)

    # âœ… Define Subplots
    apds = [mpf.make_addplot(df['ATR'], panel=1, color='red', ylabel='ATR')]

    # âœ… Plot Candlestick Chart with ATR Below
    mpf.plot(df, type='candle', volume=True, addplot=apds, style='charles', figsize=(12, 6))
```
